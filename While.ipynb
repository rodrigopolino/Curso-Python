##Estrutura While
Repetir um código de forma indeterminada até uma condição se tornar verdadeira/falsa

```python
venda = input("Registre um produto, Para cancelar o registro de um novo produto, basta apertar enter com a caixa vazia: ")
vendas = []

while venda != '':
  vendas.append(venda)
  venda = input("Registre um produto, Para cancelar o registro de um novo produto, basta apertar enter com a caixa vazia: ")

print("Registro Finalizado. Os produtos cadastrados foram: {}".format(vendas))
```

# Cuidado com o while -> Loop Infinito

Sempre que for usar o comando while, lembre-se de ter certeza que o programa vai terminar em algum momento

### Exemplo

Digamos que temos uma lista de vendedores e as quantidades vendidas e queremos identificar todos os vendedores que bateram a meta de 50 vendas.

```python
vendas = [941, 852, 783, 714, 697, 686, 685, 670, 631, 453, 386, 371, 294, 269, 259, 218, 208, 163, 125, 102, 87, 47, 7]
vendedores = ['Maria', 'José', 'Antônio', 'João', 'Francisco', 'Ana', 'Luiz', 'Paulo', 'Carlos', 'Manoel', 'Pedro', 'Francisca', 'Marcos', 'Raimundo', 'Sebastião', 'Antônia', 'Marcelo', 'Jorge', 'Márcia', 'Geraldo', 'Adriana', 'Sandra', 'Luis']
meta = 700

i = 0

while vendas[i] > meta:
  print(f"{vendedores[i]} bateu a meta. Vendas: {vendas[i]}")
  i += 1
```

Vamos criar um sistema de vendas. Nosso programa deve registrar os produtos e as quantidades (2 inputs) e adicionar em uma lista.

O programa deve continuar rodando até o input ser vazio, ou seja, o usuário apertar enter sem digitar nenhum produto ou quantidade.

Ao final do programa, ele deve printar todos os produtos e quantidades vendidas.

Obs: Caso queira, para o print ficar mais visual, pode usar o join para cada item ser printado em uma linha.
Sugestão para sua lista de produtos vendidos:

```python
vendas = [
    ['maçã', 5],
    ['banana', 15],
    ['azeite', 1],
    ['vinho', 3],
]
```

```python
vendas = []

while True:
  produto = input("Qual o produto? ")
  if not produto:
    break
  qtde = int(input("Qual é a quantidade? "))
  vendas.append([produto, qtde])

print(vendas)
```

#### 2. Faça um programa que leia um nome de usuário e a sua senha e não aceite a senha igual ao nome do usuário, mostrando uma mensagem de erro e voltando a pedir as informações.

```python
usuario = input("Digite o nome do usário: ")
senha = input("Digite um senha: ")

while senha == usuario:
  print("Erro! A senha não pode ser igual ao nome do usuário")
  usuario = input("Digite seu nome: ")
  senha = input("Digite o nome do usário: ")
print("")
print(f"Usuário: {usuario}")
print(f"Senha: {senha}")
```

#### 3. Faça um programa que leia e valide as seguintes informações (e para cada uma delas, continue pedindo a informação até o usuário inserir corretamente):
##### Nome: maior que 3 caracteres;
##### Idade: entre 0 e 150;
##### Salário: maior que zero;
##### Sexo: 'f' ou 'm';
##### Estado Civil: 's', 'c', 'v', 'd';

```python
nome = input("Digite o nome com mais de 3 caracteres: ")
while len(nome) < 4:
  print("Nome incorreto incorreto! Tente novamente")
  nome = input("\nDigite o nome com mais de 3 caracteres: ")

idade = int(input("Digite a sua idade: "))
while not (0<=idade<=150):
  print("Idade incorreta! Tente novamente")
  idade = int(input("\nDigite a sua idade: "))

salario = int(input("Digite seu salário: "))
while salario <= 0:
  print("Salário incorreto! Tente novamente")
  salario = int(input("\nDigite seu salário: "))

sexo = input("Sexo'f' ou 'm': ")
while not sexo in ['f', 'm']:
  print("Sexo incorreto! Tente novamente")
  sexo = input("\nSexo'f' ou 'm': ")

est_civil = input("Estado Civil 's', 'c', 'v', 'd': ")
while not est_civil in ['s', 'c', 'v', 'd']:
  print("\nEstado civil incorreto! Tente novamente")
  est_civil = input("Estado Civil 's', 'c', 'v', 'd': ")

print("\nDados cadastrados")
print(f"Nome: {nome} \nIdade: {idade} \nSalário: {salario} \nSexo: {sexo} \nEstado Civil: {est_civil}")
```

#### 4. Supondo que a população de um país A seja da ordem de 80000 habitantes com uma taxa anual de crescimento de 3% e que a população de B seja 200000 habitantes com uma taxa de crescimento de 1.5%. Faça um programa que calcule e escreva o número de anos necessários para que a população do país A ultrapasse ou iguale a população do país B, mantidas as taxas de crescimento.

```python
A = 80000
B = 200000

anos = 0
while A < B:
  A *= 1.03
  B *= 1.015
  anos += 1
print(f"{anos} anos necessários para que a população do país A ultrapasse ou iguale a população do país B, mantidas as taxas de crescimento")
```

#### 7. Faça um programa que peça para o usuário inserir o faturamento dos últimos 5 meses (individualmente) e informe o faturamento total (soma) e o faturamento médio por mês (média).

```python
faturamento = []
total = 0
for i in range(5):
  fat = float(input(f"Digite o faturamento do {i+1}º mês: "))
  faturamento.append(fat)
  total += fat

media_faturamento = total / len(faturamento)
print("A soma do faturamento foi de: {} e a média foi de {}".format(total, media_faturamento))
```

#### 8. Faça um programa que consiga categorizar a idade das equipes de uma empresa. Faça um programa que peça para n pessoas a sua idade, ao final o programa devera verificar se a média de idade da equipe varia entre 0 e 25 (jovem) ,26 e 60 (sênior) e maior que 60 (idosa); e então, dizer se a equipe é jovem, sênior ou idosa, conforme a média calculada.

```python
qtde = int(input("Digite quantas pessoas serão: "))

total = 0
for i in range(qtde):
  idade = int(input(f"Digite a idade da {i+1}ª pessoa:"))
  total += idade

media = total / qtde

if 0<=media<=25:
  print("Equipe Jovem")
elif 26<=media<=60:
  print("Equipe Sênior")
elif media > 60:
  print("Equipe Idosa")
```

#### 9. Numa eleição existem três candidatos. Faça um programa que peça o número total de eleitores. Peça para cada eleitor votar e ao final mostrar o número de votos de cada candidato.

```python
total = int(input("Digite o número total de eleitores: "))

soma_1 = 0
soma_2 = 0
soma_3 = 0

for i in range(total):
  voto = int(input("Digite o candito que você quer votar (1, 2 ou 3): "))
  if voto == 1:
    soma_1 += 1
  elif voto == 2:
    soma_2 += 1
  elif voto == 3:
    soma_3 += 1
  else:
    print(f"Voto do eleitor {i+1} inválido!")

print("Votos do candidato 1: {}".format(soma_1))
print("Votos do candidato 2: {}".format(soma_2))
print("Votos do candidato 3: {}".format(soma_3))
```
