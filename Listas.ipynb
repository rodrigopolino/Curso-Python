#Tamanho da lista e Maior e Menor valor

```python
produtos = ['apple tv', 'mac', 'iphone 11', 'IPad', 'apple watch', 'mac book', 'airpods']

print(len(produtos)) #função - Tamanho da lista

vendas = [1000, 1500, 15000, 270, 900, 100, 1200]

maior_venda = max(vendas) #mais vendido
menor_venda = min(vendas) #menos vendido

max = vendas.index(maior_venda)
produto_mais_vendido = produtos[max]

min = vendas.index(menor_venda)
produto_menos_vendido = produtos[min]

print("O produto mais vendido é o: {} e o produto menos vendido é o: {}".format(produto_mais_vendido, produto_menos_vendido))




```

### Exercício 3
#### Crie um sistema de consulta de bônus dos funcionários
Seu sistema deve:
- Pegar o valor de vendas do funcinoário por meio de um input
- Calcular o bônus do funcionário de acordo com a seguinte regra:
      - Se o funcionário vendeu mais de 1000 unidades, ele ganha R$2 de bonus para cada unidade vendida
      - Se o funcionário vendeu mais de 5000 unidades, ele ganha R$2 de bônus para cada unidade + um valor fixo de R$1000
      - Se o funcionário vendeu menos de 1000 unidades, ele não ganha bônus
- Printar no final o valor do bônus do funcionário


```python
vendas = int(input("Digite o valor de vendas: "))

if vendas >= 1000:
  bonus = vendas * 2
  print(f"R${bonus}")
if vendas >= 5000:
  bonus = vendas * 2 + 1000
  print(f"R${bonus}")
else:
  print("Você não ganhou bônus")
```
